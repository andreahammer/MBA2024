```python

# Example DataFrame with a column called 'SKS'

df = pd.DataFrame(diagnise_desc)

# Your substance mapping
substance_mapping = {
    'Infliximab': {
        'ML04AB02': 'L04AB02',
        'BOHJ18A1': 'L04AB02',
        'L04AB02': 'L04AB02',
    },
    'Certolizumab': {
        'ML04AB05': 'L04AB05',
        'BOHJ18A5': 'L04AB05',
        'L04AB05': 'L04AB05',
    },
    'Ustekinumab': {
        'ML04AC05': 'L04AC05',
        'BOHJ18B3': 'L04AC05',
        'L04AC05': 'L04AC05',
    }
}

# Flatten the mapping into a single dictionary
flat_mapping = {}
for substance, codes in substance_mapping.items():
    for code, reference_code in codes.items():
        flat_mapping[code] = reference_code

# Map the SKS codes to reference codes and create a new column 'reference_code'
df['reference_code'] = df['SKS'].map(flat_mapping)

# Display the updated DataFrame
df
```

```python
# Eksempel DataFrame
data = {
    'substance_code': ['kode_1a', 'kode_2b', 'kode_3c', 'kode_1b', 'kode_2a', 'kode_3a'],
    'value': [10, 20, 30, 15, 25, 35]
}
df = pd.DataFrame(data)

# Erstat koderne med referencekoden
df['reference_code'] = df['substance_code'].map(flat_mapping)
```
